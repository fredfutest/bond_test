```python
def change(filename, syear, eyear):
    df = pd.DataFrame(columns=['证券代码', '证券简称', '年份'])

    df0 = pd.read_csv(filename)
    df0.columns = map(lambda s: s.replace('\r', ''), df0.columns)
    if df0.iloc[-1, 0] == '数据来源：Wind':
        df0 = df0.iloc[:-2]

    cw = df0

    #填充证券代码和年份`#
    temp = pd.DataFrame([[a, b] for a in cw['证券代码'].unique() for b in range(syear, eyear+1)])


    df['证券代码'] = temp[0]
    df['年份'] = temp[1]

    daima2jiancheng = {key:value for key, value in cw[['证券代码', '证券简称']].values}

    df['证券简称'] = df['证券简称'].apply(lambda s: daima2jiancheng[s])

    # 各种指标
    cols = pd.Series(list(map(lambda s: s.split('\n')[0], cw.columns[2:]))).drop_duplicates().values
    cols

    for col in cols:

        for each in cw.columns:
            if each.startswith(col+'\n'):
                col2 = each
                print(each)
                break

        def get_data(ss, data):
            ticker = ss['证券代码']
            year = ss['年份']
            return data[data['证券代码']==ticker][col2.replace(str(syear), str(year))].iloc[0]

        # df[col] = df.progress_apply(lambda s: get_data(s, cw), axis=1)
        df[col] = df.apply(lambda s: get_data(s, cw), axis=1)



    df.to_excel('./整理后数据/'+filename, index=False)
```

```python
change('full_data.CSV',2011,2019)
```
